```python
import pandas as pd
from numpy import mean
import pandas as pd
import numpy as np
from numpy import std
from sklearn.datasets import make_regression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedKFold
from sklearn.ensemble import RandomForestRegressor

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
import xgboost
from xgboost import XGBRegressor
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Model, Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import random
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv('/content/Housepricedata.csv')
```

```python

y=df['SalePrice']
```

```python
x=df.iloc[:,:80]
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
fig = plt.figure(figsize =(15, 12))

# Creating plot
plt.boxplot(df.iloc[:,3]) 
# show plot
plt.show()
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
X = OneHotEncoder().fit_transform(x).toarray()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)
```

```python
from sklearn import datasets, linear_model
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats



X2 = sm.add_constant(X_train)
est = sm.OLS(y_train, X2)
est2 = est.fit()
print(est2.summary())
```

```python
linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X_train, y_train)  # perform linear regression
Y_pred = linear_regressor.predict(X_test)  # make predictions
```

```python
r2_score(y_test, Y_pred)
```

```python
rfmodel = RandomForestRegressor()
rfmodel.fit(X_train, y_train)
rfpredict=rfmodel.predict(X_test)
r2_score(y_test, rfpredict)
```

```python
import matplotlib.pyplot as plt
plt.ylabel('some numbers')
plt.plot(rfpredict,y_test,'*')
plt.show()
```

```python
import plotly.express as px
fig3 = px.scatter(x=rfpredict,y=y_test,trendline="ols")
fig3.show()
```

```python
importance = rfmodel.feature_importances_
```

```python
xgbmodel = XGBRegressor()
xgbmodel.fit(X_train, y_train)
xgbpredict=xgbmodel.predict(X_test)
r2_score(y_test, xgbpredict)
```

```python
xgbimportance =xgbmodel.feature_importances_
```

```python
import matplotlib.pyplot as plt
plt.ylabel('some numbers')
plt.plot(xgbpredict,y_test,'*')
plt.show()
```
